# 9 - Particionamento do conjunto de dados em conjunto de treino e conjunto de teste

```python
import pandas as pd
import numpy as np

df_wine = pd.read_csv('https://archive.ics.uci.edu/'
                      'ml/machine-learning-databases/wine/wine.data',
                      header=None)

# if the Wine dataset is temporarily unavailable from the
# UCI machine learning repository, un-comment the following line
# of code to load the dataset from a local path:

# df_wine = pd.read_csv('wine.data', header=None)


df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
                   'Alcalinity of ash', 'Magnesium', 'Total phenols',
                   'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                   'Color intensity', 'Hue', 'OD280/OD315 of diluted wines',
                   'Proline']

print('Class labels', np.unique(df_wine['Class label']))
df_wine.head()
```

**Comentário 01:** Na célula acima temos a importação das bibliotecas pandas (com o aliás 'pd') e numpy (com o aliás 'np'). Em seguida, `df_wine` recebe um dataframe lido a partir do método pandas `read_csv` utilizando uma `URL` (como o parâmetro `header` recebe o valor `none`, isso quer dizer que nenhuma linha de texto do cabeçalho no arquivo de origem do dataframe será excluída). Após isso, é utilizado o método `columns` para definir os rótulos do nosso dataframe e o `print('Class labels', np.unique(df_wine['Class label']))` para printar os valores exclusivos da coluna `Class label` do nosso dataframe. Por fim, o método `.head()` é utilizado para exibir as 5 primeiras linhas do nosso dataframe, partindo do topo (apenas as 5 primeiras linhas são exibidas uma vez que nenhum valor foi passado para o método).

```python
from sklearn.model_selection import train_test_split

X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

X_train, X_test, y_train, y_test =\
    train_test_split(X, y, 
                     test_size=0.3, 
                     random_state=0, 
                     stratify=y)
```

**Comentário 02:** Na célula de código acima temos a importação de `train_test_split()` da biblioteca de ciência de dados `scikit-learn`, que é utilizada para dividir um conjunto de dados em subconjuntos que minimizam o potencial de viés em seu preocesso de avaliação e validação. Em seguida, `X` recebe os valores de todas as linhas a partir da segunda coluna do dataframe `df_wine`, enquanto `y` recebe os valores de todas as linhas da primeira coluna de `df_wine` (para extrair os dados de linhas/colunas específicas do dataframe, são utilizados os métodos `.iloc[<linhas>, <colunas>]` e `.values`). Por fim, `X_train`, `X_test`, `y_train`, `y_test` recebem os valores retornados de `train_test_split`. Sigua a explicação:

- `x_train` guarda a parte de treinamento da primeira sequência - x.
-  `x_test` guarda a parte de teste da primeira sequência - x. 
- `y_train` guarda a parte de treinamento da segunda sequência - y. 
- `y_test` guarda a parte de teste da segunda sequência - y.

Parâmetros utilizados em `train_test_split`:

- `test_size=0.3` define o tamanho do conjunto de teste para 0.3%.
- `random_state` controla a randomização durante a divisão.
- `stratify` é um objeto semelhante a uma matriz que determina como usar uma divisão estratificada.
