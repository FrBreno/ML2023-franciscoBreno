# 3 - Tipos e Conversão de tipos
O Numpy suporta uma variedade de tipos muito maior do que o Python.

Existem 5 tipos básicos numéricos
- Booleans (bool)
- Integers (int)
- Unsigned Intergers (uint)
- Floating Points (float)
- Complex

Complementarmente, a célula abaixo mostra um quadro comparativos com uma maior quantidade de tipos do Numpy.

```python
import numpy as np
```

```python
x = np.float32(1.0)
x
```

**Comentário 01:** `x` recebe o valor do argumento de `.float32` convertido na faixa de valores do tipo de dado `float32`.

```python
y = np.int_([1,2,4])
y
```

**Comentário 02:** A lista `[1,2,3]` tem seus valores convertidos para o tipo de dado `ìnt`. O valor retornado é atribuído a `y`.

```python
z = np.arange(3, dtype=np.uint8)
z
```

**Comentário 03:** `z` recebe um array contendo uma sequência de 3 valores do intervalo de valores do tipo `uint8` partindo do seu primeio valor, no caso o número zero, espaçando de maneira uniforme até chegar ao terceiro valor.

```python
np.array([1, 2, 3], dtype='f')
```

## Convertendo tipos de arrays

Para convertermos arrays Numpy, basta utilizarmos o método astype().

```python
z.astype(float)
```

**Comentário 05:** O método `astype()` retorna o array convertido para o tipo especificado.

```python
np.int8(z)
```

**Comentário 06**: Na célula acima, `np.int8(z)` está retornando o array `z` convertido para o tipo de dado `int8`

Para verificarmos o tipo de um array Numpy, basta verificarmos o seu atributo dtype.

```python
z.dtype
```

**Comentário 07:** Note que nenhum dos métodos de conversão de array's mostrados acima alterou o tipo do array original.
