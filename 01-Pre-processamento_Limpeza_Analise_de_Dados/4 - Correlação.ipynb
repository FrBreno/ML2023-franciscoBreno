# 4 - Correlação
**Direto ao ponto**: correlação é o coeficiente de interdependência entre duas ou mais variáveis.

Em muitos casos, quando analisamos algumas variáveis, essa interdependência é muito clara. Exemplo:
- Massa corporal e altura,
- Consumo de combustível e distância percorrida,
- Número de anúncios e produtos vendidos,
- Tempo de estudo e desempenho na avaliação.

Entretanto, algumas vezes essa correlação pode não ser muito clara.

## Correlações positivas

```python
import numpy as np

np.random.seed(1)

# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Positive Correlation with some noise
y = x + np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

**Comentário 01:** Temos a importação da biblioteca numpy com o aliás 'np'. Em seguida é definido uma `seed` para geramos valores aleatórios com o método `random`.

`x` recebe um array com 1000 números inteiros aleatórios dentro do intervalo de 0 a 50. Enquanto `y` recebe o resultado da soma de matriz entre `x` e o array de 1000 amostras aleatórias de uma distribuição normal (dentro do intevalo de 0 a 10) gerado através do comando `np.random.normal(0, 10, 1000)`.

Ao final, temos o retorno do coeficiente de correlação entre `x` e `y` a partir do comando `np.corrcoef(x, y)`.

```python
# import matplotlib
import matplotlib.pyplot as plt

plt.scatter(x, y)
```

**Comentário 02:** Na célula de código acima temos a importação da biblioteca do matplotlib.pyplot com o aliás 'plt' e em seguida temos o plot do gráfico de dispersão a partir dos valores de `x` e `y`. Note a curva positiva dos dados.

## Correlações negativas

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Negative Correlation with some noise
y = 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

**Comentário 03:** Na célula de código acima temos que `x` recebe um array de 1000 inteiros aleatórios dentro da faixa de valores de 50 a 1000. Em seguida, temos que `y` recebe um array resultante da subtração de 100 pelos elemento do array retornado da soma de 1000 amostras aleatórias de uma distribuição normal (dentro do intevalo de 0 a 5) gerado através do comando `np.random.normal(0, 5, 1000)` com a matriz `x`.

Por fim, temos o retorno do coeficiente de correlação entre `x` e `y` a partir do comando `np.corrcoef(x, y)`.

```python
plt.scatter(x, y)
```

**Comentário 04:** Na célula de código acima temos o plot do gráfico de dispersão a partir dos valores de `x` e `y`. Note a curva negativa dos dados (representando a correlação negativa dos mesmos).

## Correlação fraca ou sem correlação

```python
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

**Comentário 05:** Na célula de código acima temos a criação de dois numpy array com 1000 inteiros aleatórios dentro de uma faixa de valores de 0 a 50.

Na sequência temos o retorno do coeficiente de correlação entre `x` e `y` a partir do comando `np.corrcoef(x, y)`.

```python
plt.scatter(x, y)
```

**Comentário 06:** Na célula de código acima temos o plot do gráfico de dispersão a partir dos valores de `x` e `y`. Note a bagunça dos dados (representando a correlação fraca ou a ausência de correlação).
