## Lendo arquivos de texto puro

```python
filename = 'huck_finn.txt'

file = open(filename, mode='r')
text = file.read()

file.close()
```

```python
print(text)
```

**Comentário 01:** Nas células de código acima temos a abertura de um arquivo `.txt` em modo de leitura (no qual o nome foi armazenado em `filename`) por meio do comando `file = open(filename, mode='r')`. Em seguida, o conteúdo do arquivo foi "lido" para dentro de `text` por meio do método `read()` e o arquivo foi fechado com `file.close()`. Por fim, temos o `print` do conteúdo de `text`.

## Criando um contexto para leitura de arquivos

```python
with open('huck_finn.txt', 'r') as file:
    print(file.read())
```

**Comentário 02:** A célula de código acima tem a mesma finalidade de execução que as células de código anteriores, só que dessa vez, criando um contexto para a abertura do arquivo com o comando `with open('huck_finn.txt', 'r') as file:`.

## Importando dados com o Numpy

```python
import numpy as np

filename = 'MNIST.txt'

data = np.loadtxt(filename, delimiter=',')
data
```

**Comentário 03:** Na célula de código acima temos a importação da biblioteca `numpy` com o aliás 'np' e a atribuição do nome do arquivo `.txt` para `filename`. Em seguida, `data` recebe um array retornando os dados da abertura do arquivo com o método numpy `loadtxt()` (note que o método recebe como parâmetro `delimiter=','` para separar os dados com a `,`).

```python
data = np.loadtxt(filename, delimiter=',', skiprows=1)
```

```python
print(data)
```

**Comentário 04:** O parâmetro `skiprows` é utilizado para dizer quantas linhas devem ser saltadas antes de iniciar a leiruta do arquivo. Nesse caso, como seu valor é 1, a primeira linha do arquivo será ignorada.

```python
data = np.loadtxt(filename, delimiter=',', skiprows=1, usecols=[0,2])
data
```

**Comentário 05:** O parâmetro `usecols` diz ao método quais colunas serão extraídas do arquivo. Nesse caso, no código `np.loadtxt(filename, delimiter=',', skiprows=1, usecols=[0,2])`, `usecols=[0,2]` diz ao método `loadtxt()` que queremos extrair os valores das colunas `0` e `2` do nosso arquivo.

```python
data = np.loadtxt(filename, delimiter=',', dtype=str)
print(data)
```

**Comentário 06:** O parâmetro `dtype` diz ao método `loadtxt()` qual será o tipo de dado da matriz resultante. Nesse caso, a matriz resultante `data` terá o tipo de dados `string` (data = np.loadtxt(filename, delimiter=' , ', `dtype=str`))

```python
import pandas as pd

filename = 'winequality-red.csv'

data = pd.read_csv(filename)

data.head()
```

**Comentário 07:** Na célula de código acima temos a importação da biblioteca `pandas` com o aliás 'pd' e a atribuição do nome de um arquivo `.csv` para `filename`. Em seguida, `data` recebe um dataframe retornado da abertura do arquivo `filename` com o método pandas `read_csv()` e temos a aplicação do método `head()` em `data`.

O método `head()` retorna um número especificado de linhas a partir do topo. Caso nenhum valor seja especificado, o método retorna as 5 primeiras linhas do dataframe.

```python
data_array = data.values
data_array
```
