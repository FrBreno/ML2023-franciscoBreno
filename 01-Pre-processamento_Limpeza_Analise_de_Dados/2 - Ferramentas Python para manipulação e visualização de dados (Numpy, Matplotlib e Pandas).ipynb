**Comentário 17:** Retorna os dados armazenados no arquivo.

```python
# Usar memory mapping quando o conteúdo não couber em memória RAM
x_mm = np.memmap('/tmp/x_mm.npy', dtype='float32', mode='w+', shape=(1000, 100))
x_mm[999][99] = 12.5
print(x_mm[999][99])
```

**Comentário 18:** O método `.memmap` cria um mapa de memória para uma matriz 1000 por 100 armazenada em um arquivo binário no disco. Nesse caso, os dados do conteúdo original são instanciados na matriz com o tipo `float32`. Note que o modo de acesso ao conteúdo original é de criar ou substituir o arquivo existente para leitura e gravação (`'w+'`);

```python
# Delete the memmap instance to close the file (the mmap file is not deleted). 
# Deletion flushes memory changes to disk before removing the object.
del(x_mm)
```

**Comentário 19:** Como foi explicado nos comentários acima, `del(x_mm)` exclui a instância memmap para que seja possível fechar o arquivo de forma correta e sem percas (as alterações são salvas no disco antes da remoção do objeto).

```python
# Load
x_mm = np.memmap('/tmp/x_mm.npy', dtype='float32', mode='r', shape=(1000, 100))
print(x_mm[999][99])
```

**Comentário 20:** Essa célula de código mostra que as alterações realizadas por meio da instância memmap não são perdidas após a realização do comando `del(x_mm)`.
